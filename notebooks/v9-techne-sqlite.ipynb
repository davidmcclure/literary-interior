```python
from lint.models import Token

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
def plot_token_series(token, figsize=(3,2)):
    
    series = Token.token_series(token)
    
    plt.figure(figsize=figsize)
    
    plt.title('"{0}"'.format(token))
    plt.xlabel('Narrative time')
    plt.ylabel('Count')
    plt.plot(*zip(*series.items()))
    
    print(token)
    plt.show()
```

```python
plot_token_series('young')
plot_token_series('boy')
plot_token_series('girl')
plot_token_series('school')

plot_token_series('death')
plot_token_series('victory')
plot_token_series('defeat')
plot_token_series('embrace')
```

```python
def plot_corpus_token_series(token, figsize=(5,2)):
    
    gail = Token.token_series(token, corpus='gail')
    chicago = Token.token_series(token, corpus='chicago')
    
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    plt.title('Offset-binned token counts')
    plt.tight_layout(w_pad=3)

    ax1.set_title('"{0}" - Gail'.format(token))
    ax1.plot(*zip(*gail.items()))

    ax2.set_title('"{0}" - Chicago'.format(token))
    ax2.plot(*zip(*chicago.items()))

    print(token)
    plt.show()
```

```python
plot_corpus_token_series('young')
plot_corpus_token_series('boy')
plot_corpus_token_series('girl')
plot_corpus_token_series('school')

plot_corpus_token_series('death')
plot_corpus_token_series('victory')
plot_corpus_token_series('defeat')
plot_corpus_token_series('embrace')
```

```python
plot_corpus_token_series('marriage')
plot_corpus_token_series('married')
plot_corpus_token_series('wife')
plot_corpus_token_series('husband')
plot_corpus_token_series('wedding')
```
