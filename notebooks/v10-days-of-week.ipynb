```python
from lint.models import Token

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
import numpy as np

from scipy.ndimage.measurements import center_of_mass

def plot_center_of_mass(token, figsize=(3,2)):
    
    series = Token.token_series(token)
    
    counts = list(series.values())
    
    center = center_of_mass(np.array(counts))[0]
    
    plt.figure(figsize=figsize)
    
    plt.title('"{0}"'.format(token))
    plt.xlabel('Narrative time')
    plt.ylabel('Count')
    plt.plot(*zip(*series.items()))
    
    print(token)
    print('Center of mass: {0}'.format(center))
    plt.show()
```

```python
plot_center_of_mass('sunday')
plot_center_of_mass('saturday')
plot_center_of_mass('friday')
plot_center_of_mass('thursday')
plot_center_of_mass('wednesday')
plot_center_of_mass('tuesday')
plot_center_of_mass('monday')
```
