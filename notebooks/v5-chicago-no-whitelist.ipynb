# Baseline series

```python
from lint.models import Offset

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
chicago = Offset.baseline_series('chicago')
    
plt.title('Chicago')
plt.xlabel('Position')
plt.ylabel('Count')
plt.plot(*zip(*chicago.items()))
plt.show()
```

Still not flat because we're throwing out non [a-z] tokens. Since I think many of the tokens that get discarded have quotes in them - eg, `"Word` - maybe this tracks dialogue? Either way - solution is to flatten the text out to ASCII before tokenizing, which should fix most of these problems.
