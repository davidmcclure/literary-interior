# Baseline series

```python
from lint.models import Offset

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
chicago = Offset.baseline_series('chicago')
    
plt.title('Chicago')
plt.xlabel('Position')
plt.ylabel('Count')
plt.plot(*zip(*chicago.items()))
plt.show()
```

# Token series

```python
def plot_token_series(token):
    
    series = Offset.token_series(token)
    
    print(token)
    plt.title('Offset-binned token counts')
    plt.xlabel('Position')
    plt.ylabel('Count')
    plt.plot(*zip(*series.items()))
    plt.show()
```

```python
plot_token_series('the')
plot_token_series('him')
plot_token_series('a')
plot_token_series('an')
```

# Token densities

```python
def plot_token_density(token):
    
    density = Offset.token_density(token)
    
    print(token)
    plt.title('Offset-binned densities')
    plt.xlabel('Position')
    plt.ylabel('Density')
    plt.plot(*zip(*density.items()))
    plt.show()
```

```python
plot_token_density('the')
plot_token_density('a')
plot_token_density('an')
```
