# Baseline series

```python
y1 = 1800
y2 = 1920
```

```python
from lint.models import Offset

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline

def plot_baseline_series(year1=y1, year2=y2):
    
    series = Offset.baseline_series(year1, year2)
    
    plt.title('All tokens')
    plt.xlabel('Position')
    plt.ylabel('Count')
    plt.plot(*zip(*series.items()))
    plt.show()
```

```python
plot_baseline_series()
```

# Raw token series

```python
def plot_token_series(token, year1=y1, year2=y2):
    
    series = Offset.token_series(token, year1, year2)
    
    print(token)
    plt.title('Offset-binned token counts')
    plt.xlabel('Position')
    plt.ylabel('Count')
    plt.plot(*zip(*series.items()))
    plt.show()
```

# Kolmogorov-Smirnov

So, we want to know - which words are significantly different from the baseline distribution for all words?

```python
from scipy import stats

baseline = Offset.baseline_series(y1, y2)
```

```python
tokens = [
    'the',
    'birth',
    'youth',
    'marriage',
    'death',
]

for token in tokens:
    
    t_series = Offset.token_series(token, y1, y2)
    
    d, p = stats.ks_2samp(
        list(baseline.values()),
        list(t_series.values()),
    )

    print(token, d, p)
```

No good - samples are so big that the test saturates before it can distinguish among words.

# Anderson-Darling

```python
tokens = [
    'the',
    'birth',
    'youth',
    'marriage',
    'death',
]

for token in tokens:
    
    t_series = Offset.token_series(token, y1, y2)
    
    a2, c, p = stats.anderson_ksamp([
        list(baseline.values()),
        list(t_series.values()),
    ])

    print(token, a2, c, p)
```

# Pearson

```python
tokens = [
    'the',
    'birth',
    'youth',
    'marriage',
    'death',
]

for token in tokens:
    
    t_series = Offset.token_series(token, y1, y2)
    
    score, p = stats.pearsonr(
        list(baseline.values()),
        list(t_series.values()),
    )

    print(token, score, p)
```

Better - let's try finding words with the lowest correlations.

```python
counts = Offset.token_counts(y1, y2)
```

```python
# Tokens with >1M occurrences.
tokens = [token for token, count in counts.items() if count > 1e7]
```

```python
corrs = []
for token in tokens:
    
    t_series = Offset.token_series(token, y1, y2)
    
    score, _ = stats.pearsonr(
        list(baseline.values()),
        list(t_series.values()),
    )
    
    corrs.append((token, score))
    
# Sort least -> most correlated.
corrs = sorted(corrs, key=lambda x: x[1])
```

```python
print(corrs[:1000])
```

So, looks like front / end matter really visible.
