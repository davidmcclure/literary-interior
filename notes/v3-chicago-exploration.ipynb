# Baseline series

```python
from lint.models import Offset

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline

def plot_baseline_series():
    
    series = Offset.baseline_series()
    
    plt.title('All tokens')
    plt.xlabel('Position')
    plt.ylabel('Count')
    plt.plot(*zip(*series.items()))
    plt.show()
```

```python
plot_baseline_series()
```

# Notes

- "the" is high at the beginning, then levels out in the middle, and then increases again at the end, I assume in some kind of climax, a re-focusing of the narrative on the concrete and external?

- "his" looks roughly like "the" - high at the beginning and end - but "him" increases monotonically (and very significantly) across the text. This seems strange and interesting. As narratives progress, men increasingly become the objects of verbs, things are increasingly _done_ to men, in some sense? By comparison, "he" - man as subject - increases very slightly, but it mostly flat. "her" - woman as object - is flat, maybe slightly declining.

- "she" looks vaguely similar to the 6-hump pattern in Hathi, if you squint? But not perfectly, and with much more noise. Not sure that anything really can be said about it here. Perhaps the sample is too small?

- "a" and "an" decrease linearly and significantly from start to end. This seems like one of the cleanest and most unambiguous trends. Generally it seems like texts begin with concretions, specific things and then move away from this. But, towards what? How does this actually pan out? What's the other end of the gradient - what's the opposite of "concrete"? Abstractions? Psychological interiority? Maybe this is just a function of the nature of beginnings - the stage needs to be set, the scene of action described, the pieces put on the board. This seems to account for the places where there are spikes at the beginning - like with "the," which levels off around 25% - but it doesn't account for things like a/an, which decrease consistently and monotonically.

- In a similar vein - numbers decrease over the course of the text, with a slightly parabolic shape - high at the beginning, fall off quickly, level out somewhat in the middle, and then fall quickly at the end. This fits well with the a/an decrease, the decline in concretion. If you label something with a count - "thee days," "two children," etc. - you tend to be talking about things that are specific, external, often physical (?).

- Likewise adjectives - descriptions of things, again I'd wager the description of the physical and external scene / setting / world - decrease consistently, with a notch upwards at the very end that looks similar to "the."
